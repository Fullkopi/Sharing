```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
mean_X1 = 3
variance_X1 = 9
mean_X2 = 4
variance_X2 = 4
n = 100

# Generate random samples
X1_samples = np.random.normal(mean_X1, np.sqrt(variance_X1), n)
X2_samples = X1_samples / 2 + np.random.normal(mean_X2, np.sqrt(variance_X2), n)

# Combine X1 and X2 into two-dimensional points
points = np.column_stack((X1_samples, X2_samples))

# Calculate the sample mean vector
sample_mean = np.mean(points, axis = 0)

# Calculate sample covariance matrix
sample_covariance_matrix = np.cov(points, rowvar=False)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(sample_covariance_matrix)

# Calculate the length of eigenvectors scaled by the square root of their corresponding eigenvalues
scaled_eigenvectors = eigenvalues[:, np.newaxis] * eigenvectors.T

# Plot points
plt.figure(figsize=(8, 8))
plt.scatter(points[:, 0], points[:, 1], label='Data Points')

# Plot eigenvectors
plt.quiver(sample_mean[0], sample_mean[1], scaled_eigenvectors[0, :][0], scaled_eigenvectors[0, :][1], color=['r'], scale_units='xy', scale=1)
plt.quiver(sample_mean[0], sample_mean[1], scaled_eigenvectors[1, :][0], scaled_eigenvectors[1, :][1], color=['b'], scale_units='xy', scale=1)
# Set plot limits and labels
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Random Data Points and Eigenvectors')
plt.legend()
plt.grid(True)

plt.show()


# Find the rotated data points
points_rotated = np.matmul(eigenvectors.T, np.transpose(points - sample_mean)).T
# Plot points
plt.figure(figsize=(8, 8))
plt.scatter(points_rotated[:, 0], points_rotated[:, 1], label='Rotated Data Points')
# Set plot limits and labels
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Rotated Data Points')
plt.legend()
plt.grid(True)

plt.show()
```
